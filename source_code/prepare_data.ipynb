```python
import pandas as pd
```

```python
data_set_list = ['../data/heart_2020_cleaned.csv',
             '../data/Noncommunicable diseases.csv',
             '../data/Cardiovascular diseases.csv',
             '../data/Chronic respiratory diseases.csv',
             '../data/Diabetes.csv',
             '../data/Obesity_unhealthy diet.csv',
             '../data/Cancer.csv',
             '../data/covid_data.csv',
             '../prepare_date/survey lung cancer.csv'
             ]
```

```python
for file_path in data_set_list:
    file_name = file_path.split('/')[-1]
    print(file_name)
    print('--------------------------')
    df = pd.read_csv(file_path)
    
    object_columns = df.select_dtypes(include=['object']).columns.tolist()
   
    for feature in object_columns:

        if all(value in ['Yes', 'No'] for value in df[feature].unique()):
            df[feature] = df[feature].fillna(0).replace('No', 0).ne(0).astype(int)

    print(df.head(2))
    df.to_csv(f'../cleaned_data/{file_name}',index=False)
    print('****************************************')
    
```
